```python
# Looking for columns that contain valuable numeric values from the table, such
# as the 'Estimated Income' column, which shows meaningful data that we can
# categorize.
```

```python
bins = [0, 100000, 300000, float('inf')]
label = ['Low', 'Medium', 'High']
df['E_income'] = pd.cut(df['Estimated Income'], bins=bins, labels=label, right=False)
df['E_income'].value_counts().plot(kind='bar')
```

```python
col_categorical = df[["Nationality","Occupation","Fee Structure","Loyalty Classification","Properties Owned","Risk Weighting" ,"BRId","GenderId","IAId"]]
for col in col_categorical:
  print (col)
  display(df[col].value_counts())
```

```python
for i , predictor in enumerate(col_categorical):
  plt.figure(i)
  sns.countplot(data=df, x=predictor,hue = "GenderId", color="red")
```
