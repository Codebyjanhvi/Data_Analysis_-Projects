```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/content/train.csv')
df.head(10)
```

```python
df.info()
```

```python
young_df = df['Age']<18
young_df.value_counts()
```

```python
df.groupby('Survived').value_counts()
```

```python
young_df_filtered = df[young_df]
survival_counts = young_df_filtered['Survived'].value_counts()
labels = ['Did Not Survive', 'Survived']
plt.title('Survival Rate of Young Passengers (< 18)')
plt.pie(survival_counts, labels=survival_counts.index, autopct='%1.1f%%',startangle=90)
plt.axis('equal')
plt.show()
```
