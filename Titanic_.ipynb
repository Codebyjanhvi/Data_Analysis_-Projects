```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/content/test.csv')
df.head(10)
```

```python
#Check for null values:

null_values= df.isnull().sum()
print(null_values)
```

```python
#Fill missing Age with mean

df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df['Age'].head(5))
```

```python
#Fill missing Embarked with mode
9
df['Embarked']= df['Embarked'].fillna(df['Embarked'].mode())
print(df['Embarked'].value_counts())
```

```python
#Drop the Cabin column (too many nulls)

df = df.drop(columns=['Cabin'])
```

Plot Visualizations (Matplotlib / Seaborn)


```python
#Plot a histogram of Age
#Histograms are for continuous numerical data (like age, height, temperature). The bars touch each other, indicating that the bins cover a continuous range.

sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution"')
plt.xlabel('Age')
plt.ylabel('Count of age ')
plt.show()
```

**Bar charts are effective for:**


1.   Comparing quantities across different groups (e.g., sales by region).
2.   Showing changes over time for discrete periods (e.g., monthly website visitors).
3.   Displaying frequency distributions for categorical variables (e.g., number of students in different majors)







```python
# Passenger Count by Sex

sns.countplot(x='Sex', data=df, color='yellow')
plt.xlabel('Sex')
plt.title('Passenger Count by Sex')
plt.ylabel('Count')
plt.show()
```

**Boxplots are excellent for:**

1.Comparing distributions between multiple groups or categories.

2.Identifying skewness and spread of data.

3.Detecting outliers at a glance.

```python
sns.boxplot(x='Pclass', y='Age', data=df, color='green')
plt.title("Age Distribution by Class")
plt.show()
```

```python
#Filter all female passengers under age 18

female_under_18 = df[(df['Sex'] == 'female') & (df['Age'] < 18)]
print(female_under_18)
female_under_18.to_csv('test.csv', index=False)
```
