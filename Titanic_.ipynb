```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('/content/test.csv')
df.head(10)
```

```python
#Check for null values:

null_values= df.isnull().sum()
print(null_values)
```

```python
#Fill missing Age with mean

df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df['Age'].head(5))
```

```python
#Fill missing Embarked with mode
9
df['Embarked']= df['Embarked'].fillna(df['Embarked'].mode())
print(df['Embarked'].value_counts())
```

```python
#Drop the Cabin column (too many nulls)

df = df.drop(columns=['Cabin'])
```

```python
#--------Geting count of sex like how female and male

df['Sex'].value_counts()
```

```python
#get Min Age and max Age

max_age=df['Age'].max()
print("Max age is :", max_age)
min_age=df['Age'].min()
print("Minmum Age is :", min_age)
```

```python
#Find average Fare by Pclass
#Find max Age for each Embarked value

print(df.groupby('Pclass')['Fare'].mean())
print(df.groupby('Embarked')['Age'].max())
```

```python
# in "Fare habing one null value we will fill this from 0 "

df['Fare'].fillna(0, inplace=True)
df.info()
```

Plot Visualizations (Matplotlib / Seaborn)


```python
#Plot a histogram of Age
#Histograms are for continuous numerical data (like age, height, temperature). The bars touch each other, indicating that the bins cover a continuous range.

sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution"')
plt.xlabel('Age')
plt.ylabel('Count of age ')
plt.show()
```

```python
# Bar Chart

sns.countplot(x='Sex', data=df, color='yellow')
plt.xlabel('Sex')
plt.title('Passenger Count by Sex')
plt.ylabel('Count')
plt.show()
```
